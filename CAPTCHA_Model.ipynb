```python
import cv2
import pandas
import seaborn
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
def threshold_img (img) :
    return cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 144, 0)

def close_img (img) :
    return cv2.morphologyEx(img, cv2.MORPH_CLOSE, np.ones((5,2), np.uint8))

def dilate_img (img) :
    return cv2.dilate(img, np.ones((2,2), np.uint8), iterations = 1)

def smooth_img (img) :
    return cv2.GaussianBlur(img, (1,1), 0)
```
